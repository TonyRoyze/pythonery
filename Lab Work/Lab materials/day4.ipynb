```r
set.seed(2)
adv <- read.csv("advertising.csv")
df <- data.frame(adv)
sample <- sample(c(TRUE, FALSE), nrow(df), replace=TRUE, prob=c(0.7,0.3))
train  <- df[sample, ]
test   <- df[!sample, ]
```

```r
plot(train$Sales, train$TV)
```

```r
ols <- lm(train$Sales ~ train$TV)
ols$coefficients
```

```r
plot(train$Sales, train$TV)
abline(ols, col = "black", lwd = 2)
```

```r
model11 <- lm(Sales ~ TV , data=train)
model12 <- lm(Sales ~ Radio , data=train)
model13 <- lm(Sales ~ Newspaper , data=train)

print(summary(model11)$adj.r.squared)
print(summary(model12)$adj.r.squared)
print(summary(model13)$adj.r.squared)
```

```r
model21 <- lm(Sales ~ TV + Newspaper , data=train)
model22 <- lm(Sales ~ TV + Radio , data=train)

print(summary(model21)$adj.r.squared)
print(summary(model22)$adj.r.squared)
# print(summary(model11)$adj.r.squared)
```

```r
model31 <- lm(Sales ~ TV + Radio + Newspaper , data=train)

print(summary(model31)$adj.r.squared)
```

```r
test_predictions <- predict(model31, newdata = test)

# Calculate the Test MSE
test_actuals <- test$Sales
test_mse <- mean((test_actuals - test_predictions)^2)

# Print the Test MSE
print(paste("Test MSE:", test_mse))
```

```r
test_predictions11 <- predict(model11, newdata = test)
test_predictions22 <- predict(model22, newdata = test)
test_predictions31 <- predict(model31, newdata = test)

# Calculate the Test MSE
test_actuals <- test$Sales
test_mse11 <- mean((test_actuals - test_predictions11)^2)
test_mse22 <- mean((test_actuals - test_predictions22)^2)
test_mse31 <- mean((test_actuals - test_predictions31)^2)

# Print the Test MSE
print(paste("Test MSE Sales ~ TV:", test_mse11))
print(paste("Test MSE Sales ~ TV + Radio:", test_mse22))
print(paste("Test MSE Sales ~ TV + Radio + Newspaper:", test_mse31))
```
