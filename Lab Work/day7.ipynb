```r
df <- datasets::mtcars
```

```r
library(dplyr)
df_numeric <- df %>% select(where(is.numeric))
```

```r
pca_result <- prcomp(df_numeric, scale. = TRUE)

summary(pca_result)
```

```r
library(ggplot2)

scree_data <- data.frame(
  Component = 1:length(pca_result$sdev),
  Variance = pca_result$sdev^2 / sum(pca_result$sdev^2)
)

ggplot(scree_data, aes(x = Component, y = Variance)) +
  geom_bar(stat = "identity", fill = "steelblue") +
  geom_line() +
  geom_point() +
  xlab("Principal Components") +
  ylab("Proportion of Variance Explained") +
  ggtitle("Scree Plot")
```

```r
# Eigenvalues

scree_data_eigenvalues <- data.frame(
  Component = 1:length(pca_result$sdev),
  Eigenvalues = pca_result$sdev^2
)

ggplot(scree_data_eigenvalues, aes(x = Component, y = Eigenvalues)) +
  geom_bar(stat = "identity", fill = "steelblue") +
  geom_line() +
  geom_abline(slope=0, intercept = 1)+
  geom_point() +
  xlab("Principal Components") +
  ylab("Eigenvalues") +
  ggtitle("Scree Plot")
```
