````python
# ii.
def vowel_count(word):
    vowels = ['a', 'e', 'i', 'o', 'u']
    count = 0
    for letter in word:
        if letter.lower() in vowels:
            count += 1
    return count

v_count = vowel_count('Statistical Programming')
v_count
````

````python
# i. 
# a)
walk = [0]

# b)
for i in range(1,1000):
    if np.random.randint(2) == 0:
        walk.append(walk[i-1] -1)
    else:
        walk.append(walk[i-1] +1)
````

````python
import matplotlib.pyplot as plt
````

````python
# ii.

plt.plot(range(0,100), walk[0:100])
plt.xlabel('Time Index')
# plt.ylabel('Random Walk')
plt.title('Random Walk simulation')
plt.show()
````

````python
# iii.

if 0 in walk[1:1000]:
    print("The walk returned to the origin at ", walk[1:1000].index(0), "th time index", sep="")
else:
    print("The walk did not return to the origin in that range")
````
