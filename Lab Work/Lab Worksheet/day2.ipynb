```r
df = read.csv("Income.csv")
```

```r
plot(df$Education, df$Income)
```

```r
slm <- lm(df$Income ~ df$Education)
```

```r
slm$coefficients
```

```r
f <- function(x) {
  slm$coefficients[1] + slm$coefficients[2] * x
}
```

```r
plot(df$Education, df$Income)
curve(f, from=10, to=22,  col = "red", lwd = 2, add = TRUE)
```

```r
lm2 <- lm(df$Income ~ poly(df$Education,3))
lm2$coefficients
```

```r
g <- function(x) {
  lm2$coefficients[1] + lm2$coefficients[2]*x + lm2$coefficients[3]*x*x + lm2$coefficients[4]*x*x*x
}
```

```r
plot(df$Education, df$Income)
```

```r
curve(g, from=10, to=22,  col = "red", lwd = 2, add = FALSE)
```
